```python
%matplotlib inline
```

```python
# Import matplotlib dependencies

import matplotlib.pyplot as plt
```

```python
# X-axis values

x_axis = ['Jan', 'Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

# Y-axis values

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create plot

plt.plot(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Plot creation plus legend label
plt.plot(x_axis, y_axis, label='Boston')

# Labels for x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')

# Set y-axis limit between 0 and 45
plt.ylim(0, 45)

# Set a title
plt.title('PyBer Fare by Month')

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()

# Limit the months to Jan-June
plt.xlim('Jan', 'June')
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2, label='Boston')

ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Month')
ax.set_ylabel('Fare($)')
ax.legend()
ax.set_ylim(0, 45)
ax.grid()
```

```python
# Basic method to create a bar chart

plt.bar(x_axis, y_axis)
```

```python
# Formatting a bar chart

plt.bar(x_axis, y_axis, color='green', label='Boston')

plt.xlabel('Month')
plt.ylabel('Fare($)')

plt.title('PyBer Fare by Month')

plt.legend()
```

```python
# Use plt.gca() to invert an axis

plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()
plt.legend()
```

```python
# Create a bar chart using the object oriented method

fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a horizontal bar chart using object oriented method

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Month')
ax.invert_yaxis()
```

```python
# Creating a scatter plot using plt.plot

plt.plot(x_axis, y_axis, 'o')
```

```python
# Creating a scatter plot with plt.scatter

plt.scatter(y_axis, x_axis, c='red', label='Chicago')
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Month')
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
# Creating a bubble chart

plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Increase bubble sizes

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Using list comprehension to replace for loop

plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, color='#87ceeb', s=[i * 5 for i in y_axis], alpha=0.8, edgecolors='black', linewidths=2, label='Boston')
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Month')
ax.set_xlim(0, 50)
ax.invert_yaxis()
ax.legend()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show
```

```python
colors = ['slateblue', 'magenta', 'lightblue', 'green', 'yellowgreen', 'greenyellow', 'yellow', 'orange', 'gold', 'indianred', 'tomato', 'mistyrose']
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, 
    explode=explode_values,
    colors=colors, 
    labels=x_axis, 
    autopct='%.1f%%')
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
explode = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
    labels=x_axis,
    autopct='%.1f%%',
    explode=explode,
    shadow=True,
    startangle=95,
    counterclock=False,
    )
```
